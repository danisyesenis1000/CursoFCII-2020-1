```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.loadtxt("datos1.txt", usecols=0)
y1 = np.loadtxt("datos1.txt", usecols=1)
y1_prima = np.loadtxt("datos1.txt", usecols=2)

x2 = np.loadtxt("datos2.txt", usecols=0)
y2 = np.loadtxt("datos2.txt", usecols=1)
y2_prima = np.loadtxt("datos2.txt", usecols=2)


x3 = np.loadtxt("datos3.txt", usecols=0)
y3 = np.loadtxt("datos3.txt", usecols=1)
y3_prima = np.loadtxt("datos3.txt", usecols=2)


x4 = np.loadtxt("datos4.txt", usecols=0)
y4 = np.loadtxt("datos4.txt", usecols=1)
y4_prima = np.loadtxt("datos4.txt", usecols=2)
```

```python
fig, a1 = plt.subplots(4, 1)
a1[0].set_title("Autofunciones", size=18)

a1[0].set_ylabel("$u(x)$", size=15)
a1[0].plot(x4, y4, "r")
a1[0].set_xlim(10.0, 20.0)
a1[0].xaxis.tick_bottom()
a1[0].xaxis.set_ticks([])
a1[0].xaxis.set_ticks_position('bottom')
a1[0].yaxis.set_ticks_position('left')

a1[1].set_ylabel("$u(x)$", size=15)
a1[1].plot(x3, y3,"r")
a1[1].set_xlim(10.0, 20.0)
a1[1].xaxis.tick_bottom()
a1[1].xaxis.set_ticks([])
a1[1].xaxis.set_ticks_position('bottom')
a1[1].yaxis.set_ticks_position('left')

a1[2].set_ylabel("$u(x)$", size=15)
a1[2].plot(x2, y2, "r")
a1[2].set_xlim(10.0, 20.0)
a1[2].xaxis.tick_bottom()
a1[2].xaxis.set_ticks([])
a1[2].xaxis.set_ticks_position('bottom')
a1[2].yaxis.set_ticks_position('left')

a1[3].set_xlabel("posicion $x$", size=15)
a1[3].set_xlim(10, 20)
a1[3].set_ylabel("$u(x)$", size=15)
a1[3].plot(x1, y1, "r")
plt.savefig("Autofunciones.png", dpi=1000, bbox_inches='tight')
```

```python
plt.figure()
# plt.plot(y1, y1_prima)
plt.title("Espacio de Fase", size=19)
plt.grid()
plt.plot(y2, y2_prima,"k-", label="$n=2$")
plt.plot(y3, y3_prima,"k-", label="$n=3$")
plt.plot(y4, y4_prima,"k-", label="$n=4$")
plt.legend()
plt.xlabel("$u$", size=18)
plt.ylabel("$\dot{u}$", size=18)
plt.savefig("Espacio_de_fase.png", dpi=1000, bbox_inches='tight')
```
